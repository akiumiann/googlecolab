```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fx(x,y,z,t,a):
  return x + (-a*x + a*y)*t
```

```python
def fy(x,y,z,t,r):
  return y + (-x*z + r*x - y)*t
```

```python
def fz(x,y,z,t,b):
  return z + (x*y - b*z)*t
```

```python
m = 2000
X_all = np.zeros(m)
Y_all = np.zeros(m)
Z_all = np.zeros(m)
T_all = np.zeros(m)

X_all[0] = 1
Y_all[0] = 1
Z_all[0] = 1

a = 10
b = 8/3
r = 28
dt = 40/m

for i in np.arange(1,m):
  X_all[i] = fx(X_all[i-1],Y_all[i-1],Z_all[i-1],dt,a)
  Y_all[i] = fy(X_all[i-1],Y_all[i-1],Z_all[i-1],dt,r)
  Z_all[i] = fz(X_all[i-1],Y_all[i-1],Z_all[i-1],dt,b)
  T_all[i] = T_all[i-1] + dt

plt.rcParams["font.size"] = 20
plt.plot(T_all, X_all, marker="", color="red")
plt.xlabel("$t$")
plt.ylabel("$x(t)$")
```

```python
plt.rcParams["font.size"] = 20
plt.plot(T_all, Y_all, marker="", color="green")
plt.xlabel("$t$")
plt.ylabel("$y(t)$")
```

```python
plt.rcParams["font.size"] = 20
plt.plot(T_all, Z_all, marker="", color="blue")
plt.xlabel("$t$")
plt.ylabel("$y(t)$")
```
