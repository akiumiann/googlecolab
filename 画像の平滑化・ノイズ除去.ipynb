```python
### 画像の平滑化・ノイズ除去 ###
import cv2
import numpy as np
from matplotlib import pyplot as plt

# 画像読み込み
img = cv2.imread('okayama-u_noise.png')

# 画像を表示
plt.figure(figsize=(15,30))
plt.subplot(5,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')

# 平均化フィルタ (カーネル3x3)
kernel3 = np.ones((3,3), np.float32)/9
img_ave3 = cv2.filter2D(img, -1, kernel3)
plt.subplot(5,2,3)
plt.imshow(cv2.cvtColor(img_ave3, cv2.COLOR_BGR2RGB))
plt.title('Average 3x3')

# 平均化フィルタ (カーネル5x5)
kernel5 = np.ones((5,5), np.float32)/25
img_ave5 = cv2.filter2D(img, -1, kernel5)
plt.subplot(5,2,4)
plt.imshow(cv2.cvtColor(img_ave5, cv2.COLOR_BGR2RGB))
plt.title('Average 5x5')

# Gaussianフィルタ (カーネル3x3)
img_gauss3 = cv2.GaussianBlur(img, (3, 3), sigmaX=1.3)
plt.subplot(5,2,5)
plt.imshow(cv2.cvtColor(img_gauss3, cv2.COLOR_BGR2RGB))
plt.title('Gaussian 3x3')

# Gaussianフィルタ (カーネル5x5)
img_gauss5 = cv2.GaussianBlur(img, (5, 5), sigmaX=1.3)
plt.subplot(5,2,6)
plt.imshow(cv2.cvtColor(img_gauss5, cv2.COLOR_BGR2RGB))
plt.title('Gaussian 5x5')

# Bilateralフィルタ (カーネル3x3)
img_bilateral3 = cv2.bilateralFilter(img, 3, sigmaColor=49, sigmaSpace=49)
plt.subplot(5,2,7)
plt.imshow(cv2.cvtColor(img_bilateral3, cv2.COLOR_BGR2RGB))
plt.title('Bilateral 3x3')

# Bilateralフィルタ (カーネル5x5)
img_bilateral5 = cv2.bilateralFilter(img, 5, sigmaColor=49, sigmaSpace=49)
plt.subplot(5,2,8)
plt.imshow(cv2.cvtColor(img_bilateral5, cv2.COLOR_BGR2RGB))
plt.title('Bilateral 5x5')

# Medianフィルタ (カーネル3x3)
img_median3 = cv2.medianBlur(img, 3)
plt.subplot(5,2,9)
plt.imshow(cv2.cvtColor(img_median3, cv2.COLOR_BGR2RGB))
plt.title('Median 3x3')

# Medianフィルタ (カーネル5x5)
img_median5 = cv2.medianBlur(img, 5)
plt.subplot(5,2,10)
plt.imshow(cv2.cvtColor(img_median5, cv2.COLOR_BGR2RGB))
plt.title('Median 5x5')
```
