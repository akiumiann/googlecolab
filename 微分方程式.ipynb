```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calc_x(x, y, dx):
  return y + np.sin(x)*dx
N = 400
x_all = np.zeros(N)
y_all = np.zeros(N)
x_all[0] = 0.0
y_all[0] = -1.0
dx = 6*np.pi / N

for i in range(1,N):
  y_all[i] = calc_x(x_all[i-1],y_all[i-1],dx)
  x_all[i] = x_all[i-1] + dx

plt.rcParams["font.size"] = 20
plt.plot(x_all, y_all, marker="", color="red")
plt.xlabel("$t$")
plt.ylabel("$x(t)$")
```
